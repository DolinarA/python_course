```python

from bs4 import BeautifulSoup
```

```python
url="https://gd.eppo.int/search?k=tommv"
print(url)
```

```python
import requests
```

```python
response = requests.get(url)
print(response)
soup = BeautifulSoup(response.content, 'html.parser')
print(soup)
```

```python
eppocodes = soup.find_all(class_="EPPOCode")
print(eppocodes)
eppocodes_list = [eppocode.text for eppocode in eppocodes]
print(eppocodes_list)
```

```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Load the webpage
driver.get(url)

# Pause for 2 seconds to allow the page to render
time.sleep(2)

# Find an element on the page using its ID
p_element = driver.find_element(By.ID, "EPPOCode")

# Print the text of the element
print(p_element.text)

# Close the browser
driver.close()
```
